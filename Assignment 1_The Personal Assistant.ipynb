```python
'''1. Ask the user for their name and store it in a variable.
2. Ask the user for their birth year (as an integer) and store it.
3. Ask the user for their current age (as an integer) and store it.
4. Calculate and print the year they will turn 100 years old.
5. Ask the user for a lucky number (as a float).
6. Print a summary message that includes all the information you've collected.
'''

name = input("Enter your name: ")   # take user's name
print("Your name is:", name)

birth_year = int(input("Enter your birth year: "))   # birth year as integer
print("Your birth year is:", birth_year)

age = int(input("Enter your current age: "))   # age as integer
print("Your age is:", age)

current_year = int(input("Enter the current year: "))

year_100 = 100 - age + current_year
print(name, "will turn 100 years old in the year", year_100)

lucky_number = float(input("Enter your lucky number: "))   # float value

print("Name:", name)
print("Birth Year:", birth_year)
print("Age:", age)
print("Lucky Number:", lucky_number)
print("You will turn 100 years old in:", year_100)
```
