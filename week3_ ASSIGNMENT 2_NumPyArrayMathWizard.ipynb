
## ASSIGNMENT 2: NumPy Array Math Wizard


```python

import numpy as np

# 2. Create arrays
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([10,20,30])

# 3. Perform arithmetic operations (broadcasting applies)
addition = arr1 + arr2
subtraction = arr1 - arr2
multiplication = arr1 * arr2
division = arr1 / arr2

print("--- Arithmetic Operations ---")
print("Addition:\n", addition)
print("Subtraction:\n", subtraction)
print("Multiplication:\n", multiplication)
print("Division:\n", division)

# 4. Broadcasting explanation
# Broadcasting automatically expands arr2 to match the shape of arr1:
# arr1 shape: (2,3)
# arr2 shape: (3,)
# arr2 is treated as [[10,20,30],[10,20,30]] during operations

# 5. Universal functions
arr3 = np.array([1,4,9,16,25])
print("\n--- Universal Functions on arr3 ---")
print("Square root:", np.sqrt(arr3))
print("Exponential:", np.exp(arr3))
print("Sine:", np.sin(arr3))

# 6. Array summary functions
arr4 = np.array([[5,12,3],[8,2,7]])
print("\n--- Array arr4 Summary ---")
print("Sum of all elements:", np.sum(arr4))
print("Max in each column:", np.max(arr4, axis=0))
print("Min in each row:", np.min(arr4, axis=1))

# 7. Type casting arr1 to float
arr1_float = arr1.astype(float)
print("\narr1 converted to float:\n", arr1_float)
```
