```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'product_a': [100, 120, 130, 150],
    'product_b': [90, 110, 120, 140],
    'product_c': [80, 100, 110, 130],
    'region': ['North', 'South', 'East', 'West']
})

# Line plot
data.set_index('month')[['product_a','product_b','product_c']].plot()
plt.title("Product Sales Over Months")
plt.show()

# Bar plot
data.plot(x='month', y='product_a', kind='bar')
plt.title("Product A Sales")
plt.show()

# Pie chart
data.groupby('region')['product_a'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.show()

# Seaborn plots
sns.distplot(data['product_a'])
plt.show()

sns.jointplot(x='product_a', y='product_b', data=data)
plt.show()

sns.boxplot(x='region', y='product_a', data=data)
plt.show()

sns.barplot(x='region', y='product_a', data=data)
plt.show()

# Heatmap
corr = data[['product_a','product_b','product_c']].corr()
sns.heatmap(corr, annot=True)
plt.show()
```
