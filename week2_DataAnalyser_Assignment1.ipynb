## Data Analyser

```python

import math
# 1. Function to get numbers from user
def get_numbers():
    user_input = input("Enter numbers separated by spaces: ")
    numbers = user_input.split()
    numbers = [float(num) for num in numbers]
    return numbers

# 2. Function to calculate mean
def calculate_mean(data):
    return sum(data) / len(data)

# 3. Function to calculate sample variance
def calculate_variance(data, mean):
    total = 0
    for num in data:
        total += (num - mean) ** 2
    variance = total / (len(data) - 1)
    return variance

# 4. Function to calculate standard deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)


# ----- Main Program -----
data = get_numbers()

mean = calculate_mean(data)
variance = calculate_variance(data, mean)
std_dev = calculate_std_dev(variance)

print("--- Statistical Analysis ---")
print("Data:", data)
print("Mean:", round(mean, 2))
print("Variance:", round(variance, 2))
print("Standard Deviation:", round(std_dev, 2))
```
