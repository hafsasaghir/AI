## Polymorphism

```python
## create a parent class animal with method sound() . create child class : dog,cat,cow each child speak override sound() method.

class Animal:
    def sound(self):
        print("animal make sound")

class dog(Animal):
    def sound(self):
        print("dog barks")
class cat(Animal):
    def sound(self):
        print("cat meows")
class cow(Animal):
    def sound(self):
        print("cow moos")
animal = [dog(), cat(), cow()]
for Animal in animal:
    Animal.sound()
```

```python
## create a class vehicle movement() method. create child class : car, bike, bus each child speak override movement() method.

class Vehicle:
    def movement(self):
        print("vehicle moves")
class car(Vehicle):
    def movement(self):
        print("car moves on road")
class bike(Vehicle):
    def movement(self):
        print("bike moves on road")
class bus(Vehicle):
    def movement(self):
        print("bus moves on road")
vehicles = [car(), bike(), bus()]
for Vehicle in vehicles:
    Vehicle.movement()

        
```

```python
## create a base class shape with method area() . create child class : circle, rectangle, square each child speak override area() method.

class Shape:
    def area(self):
        pass   # method will be overridden


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


Shapes = [Rectangle(45, 35), Circle(45), Triangle(48, 48)]

for shape in Shapes:
    print(f"Area: {shape.area()}")

```

```python
#base class employee
class employee:
    def get_role(self):
        pass
class manager(employee):
    def get_role(self):
        return "manager all overseas projects"
class developer(employee):
    def get_role(self):
        return "developer develop flutter apps"
class designer(employee):
    def get_role(self):
        return "designer design figma for apps"
Employees = [manager(), developer(), designer()]
for Employee in Employees:
    print(Employee.get_role())
    
```

```python
## payment processing system
class PaymentProcessor:
    def process_payment(self, amount):
        pass
class creditcard(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing credit card payment of {amount}")
class jazzcash(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing jazzcash payment of {amount}")
class banktransfer(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing bank transfer payment of {amount}")
payment_methods = [creditcard(), jazzcash(), banktransfer()]
for method in payment_methods:
    method.process_payment(100)
    
```
