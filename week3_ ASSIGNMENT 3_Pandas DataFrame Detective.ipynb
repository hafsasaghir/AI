```python

import pandas as pd
import numpy as np

# 2. Create dictionary
data = {
    'Name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'Age': [25,30,np.nan,35,28,np.nan],
    'Score': [85,92,78,np.nan,88,95],
    'Grade': ['A','A','B','C','B','A']
}

# 3. Create DataFrame
df = pd.DataFrame(data)
print("Initial DataFrame:\n", df)

# 4. Basic info
print("\n--- Basic Info ---")
print("First 3 rows:\n", df.head(3))
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Data types:\n", df.dtypes)

# 5. Missing data
print("\n--- Missing Data ---")
print("Is Null:\n", df.isnull())

# Fill Age with mean
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)
print("\nAge filled with mean:\n", df)

# Drop rows with missing Score
df_clean = df.dropna(subset=['Score'])
print("\nDataFrame after dropping rows with missing Score:\n", df_clean)

# 6. Selection
print("\n--- Selection ---")
# Select Name and Score
print("Name and Score columns:\n", df_clean[['Name','Score']])

# Select rows where Score > 85
high_scores = df_clean[df_clean['Score'] > 85]
print("\nRows with Score > 85:\n", high_scores)

# Set Name as index
df_indexed = df_clean.set_index('Name')
print("\nDataFrame with Name as index:\n", df_indexed)

# 7. Operations
print("\n--- Operations ---")
# Unique grades
print("Unique grades:", df_indexed['Grade'].unique())

# Grade counts
print("Grade counts:\n", df_indexed['Grade'].value_counts())

# Sort by Score descending
df_sorted = df_indexed.sort_values(by='Score', ascending=False)
print("\nDataFrame sorted by Score descending:\n", df_sorted)
```
