## nested conditionals

```python
num = 8

if num >= 0:
    print("Positive")
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
else:
    print("Negative")
```

```python
age = 22

if age >= 18:
    print("Eligible for driving")
    if age < 25:
        print("High insurance category")
    else:
        print("Normal insurance category")
else:
    print("Not eligible for driving")
```

```python
score = 85

if score >= 50:
    print("Pass")
    if score >= 75:
        print("Distinction")
else:
    print("Fail")
```

```python
num = 30

if 10 <= num <= 50:
    print("Number is in range")
    if num % 5 == 0:
        print("Divisible by 5")
else:
    print("Number is out of range")
```

```python
temperature = 28
is_raining = True

if temperature > 25:
    print("Warm weather")
    if is_raining:
        print("Carry an umbrella")
else:
    print("Cool weather")
```

```python
salary = 60000

if salary > 50000:
    print("Tax applicable")
    if salary > 100000:
        print("High tax slab")
    else:
        print("Medium tax slab")
else:
    print("No tax")
```

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Incorrect password")
else:
    print("Username not found")
```

```python
marks = 88

if marks >= 60:
    print("Eligible for admission")
    if marks >= 85:
        print("Eligible for scholarship")
    else:
        print("Not eligible for scholarship")
else:
    print("Not eligible for admission")
```
